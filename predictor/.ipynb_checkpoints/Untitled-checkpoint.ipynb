```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tsla = pd.read_csv("TSLA.csv")
tsla.head()
```

```python
tsla_mean = tsla.groupby('Date').mean()
tsla_mean.head()
```

```python
tsla_mean.plot(figsize=(12, 18))
```

```python
maximum_drop_tsla = max(tsla['Open'] - tsla['Low'])
maximum_drop_tsla
```

```python
tsla.loc[(tsla['Open'] - tsla['Low']) == maximum_drop_tsla]
```

```python
maximum_up_tsla = max(tsla['High'] - tsla['Open'])
maximum_up_tsla
```

```python
tsla.loc[(tsla['High'] - tsla['Open'] == maximum_up_tsla)]
```
